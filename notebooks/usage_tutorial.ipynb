```python
from interpolator import interpolations, interpolate
```

```python
# interpolate : Interpolates between two coordinates

# t = 0.5 : mid
# t = 0   : xyz_1
# t = 1   : xyz_2

mid_xyz = interpolate(xyz_1, xyz_2, t=0.5)
print(mid_xyz)
```

```python
# interpolations : returns interpolated images ( n = num_images )

interpolations(xyz_1, xyz_2, num_images=10)
```

```python
# interpolations with ( get_single_string = True ) : returns traj string

traj = interpolations(xyz_1, xyz_2, num_images=10, get_single_string=True)

print(traj)
```
